```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
words = ["I", "love", "deep", "learning"]
words
```

```python
attention = np.array([
    [0.10, 0.20, 0.30, 0.40],
    [0.20, 0.30, 0.30, 0.20],
    [0.30, 0.30, 0.20, 0.20],
    [0.25, 0.25, 0.25, 0.25]
])

attention
```

```python
attention.sum(axis=1)
```

```python
plt.figure(figsize=(7, 5))
sns.heatmap(
    attention,
    xticklabels=words,
    yticklabels=words,
    annot=True,
    fmt='.2f',
    cmap='YlOrRd'
)
plt.title('Attention Heatmap')
plt.xlabel('Key (Words being attended to)')
plt.ylabel('Query (Current word)')
plt.tight_layout()
plt.show()
```

```python
for i, query_word in enumerate(words):
  max_index = np.argmax(attention[i])
  print(f'Query Word : "{query_word}" attends most to "{words[max_index]}" with score {attention[i, max_index]:.2f}')
```

```python
words2 = ["Attention", "helps", "models", "focus"]

attention2 = np.array([
    [0.40, 0.20, 0.20, 0.20],
    [0.10, 0.50, 0.20, 0.20],
    [0.15, 0.20, 0.45, 0.20],
    [0.10, 0.20, 0.30, 0.40]
])

plt.figure(figsize=(7, 5))
sns.heatmap(
    attention2,
    xticklabels=words2,
    yticklabels=words2,
    annot=True,
    fmt='.2f',
    cmap='Reds'
)
plt.title('Attention Heatmap -> Example 2')
plt.xlabel('Key')
plt.ylabel('Query')
plt.tight_layout()
plt.show()
```
